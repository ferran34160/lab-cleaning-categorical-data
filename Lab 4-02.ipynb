```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
customer_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
continuous_df = customer_df.select_dtypes(float)
discrete_df = customer_df.select_dtypes(exclude=['float', 'object','datetime'])
```

```python
# There is no big correlation between variables

sns.heatmap(round(discrete_df.corr(),2), annot=True)
plt.show()

sns.heatmap(round(continuous_df.corr(),2), annot=True)
plt.show()
```

```python
def plot_discrete (data):
    for column in data:
        sns.displot(data[column])
        
plot_discrete(discrete_df)
```

```python
def plot_continuous (data):
    for column in data:
        sns.displot(data[column])
        
plot_continuous(continuous_df)
```

```python
sns.boxplot(y=continuous_df['Customer Lifetime Value'])
plt.show()
```

```python
sns.boxplot(y=continuous_df['Total Claim Amount'])
plt.show()
```

```python
# we could use this code to select some of them, but I dont think it would be a good idea to delete them
iqr = np.percentile(continuous_df['Customer Lifetime Value'],75) - np.percentile(continuous_df['Customer Lifetime Value'],25)
upper_limit = np.percentile(continuous_df['Customer Lifetime Value'],75) + 2.5*iqr
print(upper_limit)

extraordinary_points = continuous_df[continuous_df['Customer Lifetime Value'] > upper_limit]
len(extraordinary_points)  
```

```python
iqr = np.percentile(continuous_df['Total Claim Amount'],75) - np.percentile(continuous_df['Total Claim Amount'],25)
upper_limit = np.percentile(continuous_df['Total Claim Amount'],75) + 1.5*iqr
print(upper_limit)

extraordinary_points = continuous_df[continuous_df['Total Claim Amount'] > upper_limit]
len(extraordinary_points)  
```
